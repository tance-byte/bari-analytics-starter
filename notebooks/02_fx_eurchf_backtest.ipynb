# EUR/CHF — mini backtest SMA


**Objectif**: charger `data/raw/eurchf.csv` et backtester une stratégie simple **SMA 50/200** sur EUR/CHF.

Étapes suggérées:
1. Charger les données et nettoyer (dates, colonnes).
2. Calculer SMA50 et SMA200.
3. Signal: `long` si SMA50 > SMA200, sinon `cash`.
4. Calculer les rendements (log/arithm.) et la performance cumulée de la stratégie vs buy&hold.
5. Mesures: drawdown max, volatilité annuelle, ratio Sharpe (simple).
6. Tracer les courbes et exporter avec `savefig`.
7. Écrire une courte conclusion (limites, pistes d'amélioration).
