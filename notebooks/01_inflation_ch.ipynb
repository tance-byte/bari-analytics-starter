# Inflation suisse — analyse simple


**Objectif**: charger `data/raw/cpi_ch.csv`, calculer l'inflation **MoM** et **YoY**, tracer les courbes et exporter les figures.

Étapes suggérées:
1. Charger le CSV avec `pandas` (voir `src/data.py`).
2. Parser la colonne date en `datetime` et indexer par date.
3. Calculer les variations mensuelles (pct_change) et annuelles (pct_change(12)).
4. Tracer: IPC, inflation MoM, inflation YoY (utiliser `matplotlib`).
5. Exporter les graphiques via `src/utils.py::savefig`.
6. Écrire 5–10 lignes de commentaire en fin de notebook (ce que tu observes).
